**Se importan las librerias necesarias**

```python
from matplotlib import pyplot as plt
import numpy as np
```

**Funciones necesarias para calcular la probabilidad de éxito**

```python
# Retorna la probabilidad de exito si se cambia de caja en funcion N
def probSuccess(N):
    if N < 3:
        return 'El número de cajas tiene que ser mayor a 3'
    # (1/N) probabilidad de tener el premio de cada caja
    # (1/N)*(N-1) probabilidad acomulada de las cajas restantes
    # ((1/N)*(N-1))/(N-2) probabilidad de cada caja (si se abre otra caja)
    return ((1/N)*(N-1))/(N-2)
```

```python
# Retorna la probabilidad de exito de 3 cajas hasta llegar a N cajas
def probAllSuccess(N):    
    # Arreglo con numero de cajas
    ns = np.arange(3,N+1)
    # Se calcula la probabilidad de exito si hay cambio en funcion de N
    p = []
    for n in ns:
        p.append(probSuccess(n))
    return ns,p
```

```python
# Grafica la probabilidad de exito al cambiar de caja (3 cajas hasta N cajas)
def plotProbabBoxes(N):
    ns, p = probAllSuccess(N)
    plt.figure(figsize=(20,12))
    plt.plot(ns,p, 'co', linewidth=3);
    plt.yticks(np.linspace(0,p[0],20));
    plt.xticks(np.arange(3,N+1,N//21))
    plt.xlabel('Numero de Cajas (N)', fontdict={'size':16},labelpad=10)
    plt.ylabel('Probabilidad de éxito si cambio',fontdict={'size':16},labelpad=10)
    plt.grid(True)
```

**¿Cuál es la probabilidad de que, si cambio, obtenga el premio, en función de N (número de cajas)?**

```python
plotProbabBoxes(25)
```
